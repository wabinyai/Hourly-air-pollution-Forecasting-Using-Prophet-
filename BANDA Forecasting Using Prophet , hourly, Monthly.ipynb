```python
import numpy as np
import pandas as pd
from prophet import Prophet

from sklearn.metrics import mean_squared_error, mean_absolute_error

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid', {'axes.facecolor': '.9'})
sns.set_palette(palette='deep')
sns_c = sns.color_palette(palette='deep')
%matplotlib inline

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
foo =r'C:\Users\ACER\Desktop\python project\airqo/BandaHourlyPRO.csv'
df = pd.read_csv(foo, parse_dates=[0]) # We set the index column and know it has dates
```

# Hourly forecast

```python
m = Prophet(changepoint_prior_scale=0.01).fit(df)
future = m.make_future_dataframe(periods=300, freq='H')
fcst = m.predict(future)
fig = m.plot(fcst)
plt.xticks(rotation=45)
plt.xlim('2021-12-12  00:00:00','2021-12-12  23:00:00')
plt.ylim(-12,120)
plt.legend()
plot = plt.suptitle('HOURLY FORE CAST')
```

```python
m = Prophet(seasonality_mode='multiplicative').fit(df)
future = m.make_future_dataframe(periods=3652)
fcst = m.predict(future)
fig = m.plot(fcst)
plt.xlim('2021-12-12  00:00:00','2021-12-12  23:00:00')
```

```python
 fig = m.plot_components(fcst)
```
